```python
import pandas as pd
```

```python
mesh_sets = pd.read_csv("../mesh_sets.tsv", sep='\t', header=None)
```

```python
# Deals with Mesh
def format_mesh_member(input_string):
    numbers = ""
    for char in reversed(input_string):
        if char.isdigit():
            numbers = char + numbers
        else:
            break
    return "MESH:D" + numbers

new_rows = []
for index, row in mesh_sets.iterrows():
    new_rows.append(['Disease', format_mesh_member(row[1])])
    set_members = row[2].split(';')
    for member in set_members:
        formatted_member = format_mesh_member(member)
        new_rows.append(['Disease', formatted_member])
mesh_sets_pb= pd.DataFrame(new_rows)
mesh_sets_pb
```

```python
mesh_sets_pb.to_csv("mesh_sets_pb.csv")
```

```python
kegg_genesets = pd.read_csv("../kegg_geneset.tsv",  sep='\t', header=None)
pid_genesets = pd.read_csv("../pid_geneset.tsv",  sep='\t', header=None)
bio_genesets = pd.read_csv("../bio_geneset.tsv",  sep='\t', header=None)
bp_genesets = pd.read_csv("../bp_geneset.tsv",  sep='\t', header=None)
genesets=[kegg_genesets,pid_genesets,bio_genesets,bp_genesets]
```

```python
# Deals with Genes
def formart_genesets(geneset):
    def format_gene(input_string):
        numbers = ""
        for char in reversed(input_string):
            if char.isdigit():
                numbers = char + numbers
            else:
                break
        return numbers

    new_rows = []
    for index, row in geneset.iterrows():
        new_rows.append(['Gene', format_gene(row[1])])
        set_members = row[2].split(';')
        for member in set_members:
            formatted_member = format_gene(member)
            new_rows.append(['Gene', formatted_member])
    geneset_df = pd.DataFrame(new_rows)
    return geneset_df
```

```python
kegg_genesets_new = formart_genesets(kegg_genesets)
pid_genesets_new = formart_genesets(pid_genesets)
bio_genesets_new = formart_genesets(bio_genesets)
bp_genesets_new = formart_genesets(bp_genesets)
```

```python
dataframes_to_concat = [kegg_genesets_new, pid_genesets_new, bio_genesets_new, bp_genesets_new]
gene_sets_pb = pd.concat(dataframes_to_concat, ignore_index=True)
gene_sets_pb
```

```python
gene_sets_pb.to_csv("gene_sets_pb.csv")
```

```python
# All concepts in single df
all_concepts_df= pd.concat([mesh_sets_pb,gene_sets_pb], ignore_index=True)
all_concepts_df
```

```python
all_concepts_df.to_csv("all_concepts.csv")
```
